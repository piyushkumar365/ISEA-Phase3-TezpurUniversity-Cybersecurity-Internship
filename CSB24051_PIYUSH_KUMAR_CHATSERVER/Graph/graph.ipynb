```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\harsh\OneDrive-piyush\OneDrive\Desktop\ISEA\CSB24051_PIYUSH_KUMAR_CHATSERVER\performance_results.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(
    df["clients"],
    df["avg_delivery_time_ms"],
    marker="o"
)
plt.title("Number of Clients vs Average Delivery Time")
plt.xlabel("Number of Clients")
plt.ylabel("Average Delivery Time (ms)")
plt.grid(True)
plt.savefig("clients_vs_delay.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(
    df["clients"],
    df["throughput_msgs_per_sec"],
    marker="o",color='red'
)
plt.title("Number of Clients vs Throughput")
plt.xlabel("Number of Clients")
plt.ylabel("Throughput (messages/sec)")
plt.grid(True)
plt.savefig("clients_vs_throughput.png")
plt.show()
```
